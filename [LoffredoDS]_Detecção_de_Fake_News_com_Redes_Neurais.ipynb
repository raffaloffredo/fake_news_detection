# Tratamento dos Dados

Nesse projeto estamos lidando com um problema de linguagem natural, ou seja, os dados esto em formato do tipo `string`. Por esse motivo  necessrio conhecimento em _NLP_ (Processamento de Linguagem Natural) para poder tratar os dados adequadamente e obter os resultados desejados.

Com os dados padronizados podemos realizar algumas anlises complementares, como a _WordCloud_ e anlise dos dados categricos, que nesse caso iremos olhar para a autoria das notcias.

## Padronizao das palavras

Para isso ser criada a funo `clean_text` para fazer a padronizao dos textos das notcias.

```python
# criar a funo 'clean_text' que recebe o argumento 'text'.
def clean_text(text):
    # criar varivel 'word' e dividir o texto por palavras e usar espaos em branco como delimitador
    words = str(text).split()

    # converter palavras em letras minsculas, adicionando espao ao final de cada palavra
    words = [i.lower() + " " for i in words]

    # unir palavras em uma nica string, mas separadas por espaos.
    words = " ".join(words)

    # remover pontuao das strings, usando o mtodo st.punctutation como argumento
    words = words.translate(words.maketrans('', '', st.punctuation))

    return words

# aplicar a funo 'clean_text' nos atributos: 'title', 'author' e 'text'
df_clean.title  = df_clean.title.apply(clean_text)
df_clean.author = df_clean.author.apply(clean_text)
df_clean.text   = df_clean.text.apply(clean_text)

# verificar as 5 primeiras entradas para checar as alteraes feitas
df_clean.head()
```

# Mtricas de Avaliao

Antes de iniciar a construo de um modelo de aprendizado de mquina,  importante definir como iremos avaliar o modelo. Isto , decidir se ele est ou no performando como se espera.

Nesse estudo, irei utilizar 2 mtricas: a acurcia e o recall e a preciso. Com maior nfase na acurcia e no recall.

A acurcia  uma mtrica mais simples pois nos mostra um panorma geral com o resultado da quantidade de previses corretas do total de previses feitas. Sua frmula  dada por:

<br/>

$${\text{Acurcia: }}\frac{\text{qauntidade de previses corretas}} {\text{quantidade de previses realizadas}}$$

<br/>

Contudo, a acurcria no  melhor mtrica para se olhar nos casos em que os dados so desbalanceados, por exemplo: fraudes bancrias e deteco de doenas. Nesses casos, esperamos que a fraude e a pessoa doente seja exceo e no a regra. Isso naturalmente torna os dados desbalanceados. E esse pode ser o noso caso aqui. Apesar de termos visto que o conjunto de dados est devidamente balanceado, na vida real esperamos nos deparar na maior parte das vezes com notcias confivels. Logo, _fake news_ devem ser a exceo e no a regra.

Nesses casos, geralemnte utilizamos o recall como mtrica de avaliao e  por esse motivo que tambm iremos olhar para esses valores aqui ensse estudo.

# Desenvolvimento do Algoritmo utilizando TensorFlow

Para conseguirmos desenvolver o melhor algoritmo que atenda ao objetivo desse projeto, precisamos realizar um pr-processamento dos dados. Esse etapa  importante para simplificar o texto, tornando-o mais fcil de ser processado e treinado. Para isso utiliza-se a tcnica de tokenizao e de preenchimento conhecidos por _tokenizer_ e _padding_, respectivamente.

A tokenizao  uma etapa que ir reduzir o texto em unidades menores como grupo de palavras, palavras individuais, caracteres, pontuao ou smbolos. Essa unidade recebe o nome tcnico de _tokens_. Veja na imagem abaixo uma representao de tokenizao do texto em tokens, onde a unidade.

A partir disso, cada _token_ recebe um valor numrico que o representa, e isso fica salvado em forma de dicionrio, ou seja, possui as informaes de chave, no caso o _token_ e valor, a sua identificao numrica.

A etapa de _padding_, como seu prprio nome indica, ou seja, 'preenchimento', visa colocar um padro de comprimento das sequncias. Isto , todas tero o mesmo tamanho para facilitar o treinamento do modelo. Para isso,  determinado um tamanho mximo e informamos qual valor deve ser inserido nos casos que a sequncia seja menor do que o valor mximo informado.

```python
# configuraes
vocab_size = 10000  # quantidade mxima de palavras tokenizadas
trunc_type = "post" # configurar truncate
pad_type = "post"   # configurar preeenchimento
oov_tok = "<OOV>"   # token padro para tokenizao fora do vocabulrio

# tokenizar os dados de treino
tokenizer = Tokenizer(num_words = vocab_size,
                      oov_token = oov_tok)
tokenizer.fit_on_texts(train.text)

# guardar o index dos dados de treino em uma varivel
word_index = tokenizer.word_index

# aplicar tokenizing e padding
training_sequences = tokenizer.texts_to_sequences(np.array(train.text))
training_padded = pad_sequences(training_sequences,
                                truncating = trunc_type,
                                padding = pad_type)

# definir comprimento mximo de preenchimento
max_length = len(training_padded[0])

# codificar a sequncia dos dados de validao
validation_sequences = tokenizer.texts_to_sequences(np.array(validation.text))

# aplicar padding  aos dados de validao
validation_padded = pad_sequences(validation_sequences,
                                  padding = pad_type,
                                  truncating = trunc_type,
                                  maxlen = max_length)

# criar arrays com os inputs
x_train = np.copy(training_padded)
x_val   = np.copy(validation_padded)
y_train = train['label'].values
y_val   = validation['label'].values
```

Agora, com os dados devidamente preparados para serem utilizados para a construo do algoritmo, vamos cri-lo!

O modelo foi desenvolvido especificamente para o problema em questo, ou seja, determinar se uma notcia  ou no confivel. Logo, temos um problema de classificao binria. Para isso utilizou-se de uma rede neural sequencial e empilhada por camadas.

Ao todo temos 4 camadas:

<br/>

1. _Embedding_

  Cria a camada de incorporao do modelo, onde determina-se o tamanho do vocaulrio a ser utilizado, bem como o tamanho mximo da sequncia. Com isso, busca-se dar valores prximos aos tokens que tiverem significado semelhantes.

<br/>

2. _Conv1D_

  Essa  a camada de convoluo que aplica o aprendizado do modelo em lotes, com o objetivo de melhor identificar as caractersticas das sequncias, onde foram configurados que se faa use de 16 filtros de tamanho 5, bem como que se use o modelo _relu_ ( ReLU - _Rectified Linear Unit_)para ativao dos neurnios.

<br/>

3. _GlobalMaxPool1D_

  Cria a camada de _pooling_ global, ou seja, para que seja retornado apenas o valor mximo de cada lote para com isso enfatizar ainda mais as caractersticas das sequncias e reduzir a dimenso de sada.

<br/>

4. _Dense_
  Por fim, se tem a criao da camada densa, onde foi configurada com 1 unidade e ativada pela funo sigmoigal, que retorna valores booleanos 0 ou 1 para determinar a confiabilidade da notcia.

<br/>

Cabe ressaltar que foi feito um RandomSearch para encontrarmos os melhores parmetros para o nosso modelo em especfico. Por se tratar de um processo que demanda tempo e fora computacional, o mesmo foi realizado em um notebook  parte desse projeto para que aqui ficasse mais prtico e direto nessa etapa.

A seguir, se tem a construo, o treino e a impresso das caractersticas do modelo final.

```python
# construo do modelo de rede neural de classificao binria
model = tf.keras.Sequential([ # criar um modelo de rede neural sequencial empilhado por camadas
        tf.keras.layers.Embedding(vocab_size, (155), input_length=max_length), # criar camada de incoporao para dar valores prximos aos tokens que forem semelhantes
        tf.keras.layers.Conv1D(16, 5, activation='relu'), # criar camada de convoluo para aprendizado em lotes para melhor identificar caractersticas
        tf.keras.layers.GlobalMaxPooling1D(), # criar camada de pooling global par retornar somente o valor mximo de cada lote para enfatizar ainda mais as caractersticas
        tf.keras.layers.Dense(1, activation='sigmoid') # criar camada densa com uma unidade e aplicar a ativao pela funo sigmoidal
        ])

# compilar o modelo
model.compile(loss='binary_crossentropy', # definir a funo de perda
              optimizer=Adam(learning_rate=.001), # definir otimizador e taxa de aprendizado
              metrics=['accuracy','Recall','Precision','FalseNegatives']) # definir mtricas para avaliao do modelo durante o treinamento e testes
```

```python
# treinar o modelo
## verbose=2 para mostrar uma mensagem por epoch
## epochs=4 define a quantidade de pocas que o modelo passar por todo o conjunto de treinamento
history = model.fit(x_train, y_train, verbose = 2, epochs = 4,
                    validation_data = (x_val, y_val), # dados de validao
                    callbacks=[tf.keras.callbacks.EarlyStopping('val_loss', patience=3)]) # interromper o treinamento quando no houver melhoria no 'val_loss' por 3 epochs consecutivos
```
